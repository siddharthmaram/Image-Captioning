```python
from transformers import AutoModelForCausalLM
from PIL import Image
import torch
import os
from tqdm import tqdm

# Load the model
model = AutoModelForCausalLM.from_pretrained(
    "vikhyatk/moondream2",
    trust_remote_code=True,
    # dtype=torch.bfloat16,
    device_map="cuda",
)


IMG_DIRECTORY = "/kaggle/input/biology-test/test"
images = os.listdir(IMG_DIRECTORY)

preds = []

for img in tqdm(images):
    if not (img.endswith("png") or img.endswith("jpg") or img.endswith("jpeg")):
        continue
    img_path = os.path.join(IMG_DIRECTORY, img)
    # Load your image
    image = Image.open(img_path)
    
    # Optionally set sampling settings
    settings = {"temperature": 0.5, "max_tokens": 768, "top_p": 0.3}
    
    # Generate a short caption
    long_result = model.caption(
        image, 
        length="normal", 
        settings=settings
    )
    preds.append([img, long_result])
```

```python
preds_2 = [[img, d["caption"]] for img, d in preds]
```

```python
import csv

with open("preds.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["image", "caption"])
    writer.writerows(preds_2)
```
